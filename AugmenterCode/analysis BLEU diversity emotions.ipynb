```python
import sacrebleu
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
```

```python
# read the augmented data csv, as an example EDA
arq =  'covidworry_edaaug_5.csv'
eda = pd.read_csv(arq)
```

```python
eda['BLEUscore'] = eda.progress_apply(lambda x: sacrebleu.sentence_bleu(x['text'], [x['paraphrase']]), axis=1)
```

```python
eda['BLEUscore_without_bp'] = eda['BLEUscore'].progress_apply(lambda x: x.score/x.bp)
```

```python
import numpy as np
print ('eda:', np.mean(eda['BLEUscore_without_bp']))
```
